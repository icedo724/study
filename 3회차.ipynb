```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_1.csv')
#print(df.head(2))
df_q1 = df.sort_values(by='lotSizeSqFt', ascending=False).reset_index(drop=True)
rpmin = df_q1.loc[:9, 'lotSizeSqFt'].min()
df_q1.loc[:9, 'lotSizeSqFt'] = rpmin
print(round(df_q1[df_q1['yearBuilt'] >= 2000]['lotSizeSqFt'].mean(),0))

abs1 = df_q1['numOfBathrooms'].std()
df_q1['numOfBathrooms'] = df_q1['numOfBathrooms'].fillna(df_q1['numOfBathrooms'].median())
abs2 = df_q1['numOfBathrooms'].std()
print(round(abs(abs1 - abs2),3))

out_max = df['MedianStudentsPerTeacher'].mean() + (1.5 * df['MedianStudentsPerTeacher'].std())
out_min = df['MedianStudentsPerTeacher'].mean() - (1.5 * df['MedianStudentsPerTeacher'].std())
print(len(df[(df['MedianStudentsPerTeacher'] > out_max) | (df['MedianStudentsPerTeacher'] < out_min)]))
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_trainX.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_trainy.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_2_testX.csv')

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

y_train = y_train['Outcome']
co = pd.concat([x_train, test], axis=0)
scaler_target = ['Glucose', 'BloodPressure', 'Insulin']
mm = MinMaxScaler()
ss = StandardScaler()
#co[scaler_target] = mm.fit_transform(co[scaler_target])
#co[scaler_target] = ss.fit_transform(co[scaler_target])
x_train = co.iloc[:len(x_train), :]
x_test = co.iloc[len(x_train):, :]

x_t, x_va, y_t, y_va = train_test_split(x_train, y_train, test_size = 0.2, random_state = 42)

model = RandomForestClassifier(
    n_estimators = 1000,
    max_depth = 5,
    min_samples_split = 7,
    min_samples_leaf = 6,
    random_state = 42,
    class_weight = 'balanced'
)
model.fit(x_t, y_t)
pred = model.predict(x_va)
print('F1 Score:', f1_score(y_va, pred))

result = pd.DataFrame()
result['pred'] = model.predict(x_test)
print(result.isnull().sum())
print(result.head(3))
#result.to_csv('result.csv', index=False)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_3_1.csv')
print(round(df.score.mean(), 2), round(df.score.std(), 2))

from scipy.stats import ttest_1samp
st, p = ttest_1samp(df['score'], 75,alternative = 'greater')
print(round(st, 2))
print(round(p, 2))
if p < 0.05:
    print('기각')
else:
    print('채택')

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/exam/3_3_2.csv')
print(df.head(2))
from scipy.stats import chi2_contingency
ct = pd.crosstab(df['gender'], df['club_membership'])
st, p, dof, ex_ary = chi2_contingency(ct, correction=False)
print(round(st, 2))
print(round(p, 2))
if p < 0.05:
    print('기각')
else:
    print('채택')
```
