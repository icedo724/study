```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e2_p1_1.csv')
df.head(2)
```

```python
#주어진 Dataset에서 CRIM값이 가장 큰 10개의 지역을 구하고
#10개의 지역의 CRIM값을 그 중 가장 작은 값으로 대체하라. 그리고
#AGE 컬럼 값이 80이상인 대체 된 CRIM 평균값을 구하라

df = df.sort_values(by='CRIM', ascending=False).reset_index(drop=True)
df.loc[:9, 'CRIM'] = df.loc[:9, 'CRIM'].min()
an1 = df[df.AGE >= 80].CRIM.mean()
print(an1)
```

```python
#1-1에서 사용한 데이터에서 RM 중앙값으로 해당 컬럼의 결측치를 대체하라
#그리고 해당 컬럼의 결측치 대치 전후의 표준편차 차이의 절댓값을 소숫점 이하 3째자리 까지 구하라

before_RM = df.RM.std()
df.loc[df.RM.isnull(), 'RM'] = df.RM.median()
after_RM = df.RM.std()
an2 = abs(round(before_RM - after_RM, 3))
print(an2)
```

```python
#주어진 Dataset의 DIS 평균으로 부터 1.5 * 표준편차를 벗어나는 영역을 이상치라고 판단하고
#DIS 컬럼의 이상치들의 합을 구하여라.

up_cut = df.DIS.mean()+1.5*df.DIS.std()
down_cut = df.DIS.mean()-1.5*df.DIS.std()
an3 = df[df.DIS > up_cut].DIS.sum() + df[df.DIS < down_cut].DIS.sum()
print(an3)
```

```python
#순유입인원은 초중고 도내,도외 전입인원에서 초중고 도내, 도외 전출인원을 뺀값이다. 
#각년도별로 가장 큰 순유입인원을 가진 지역구의 순유입인원을 구하고 전체 기간의 해당 순유입인원들의 합을 구하여라

df['유입'] = df['초등학교_전입_도내'] + df['초등학교_전입_도외'] + df['중학교_전입_도내'] + df['중학교_전입_도외'] + df['고등학교_전입_도내'] + df['고등학교_전입_도외']
df['유출'] = df['초등학교_전출_도내'] + df['초등학교_전출_도외'] + df['중학교_전출_도내'] + df['중학교_전출_도외'] + df['고등학교_전출_도내'] + df['고등학교_전출_도외']
df['순유입'] = df['유입'] - df['유출']
an3 = df.groupby('년도')['순유입'].max().sum()
print(an3)
```

```python
# 유입, 유출 컬럼 자동 선택
inflow_cols = df.filter(like='전입').columns
outflow_cols = df.filter(like='전출').columns

# 유입, 유출, 순유입 계산
df['유입'] = df[inflow_cols].sum(axis=1)
df['유출'] = df[outflow_cols].sum(axis=1)
df['순유입'] = df['유입'] - df['유출']

# 연도별 최대 순유입 합계
an3 = df.groupby('년도')['순유입'].max().sum()
print(an3)
```

```python
#5대 범죄(절도, 사기, 배임, 방화, 폭행)의 월별 총 발생건수를 총범죄수라고 표현하자. 
#18,19년의 각각 분기별 총범죄수의 월평균 값을 구했을때 최대값을 가지는 년도와 분기를 구하고 
#해당 분기의 최댓값의 사기가 발생한 월의 사기 발생건수를 출력하라
#(1분기:1,2,3월 / 2분기 : 4,5,6월 / 3분기 7,8,9월 / 4분기 10,11,12월 , 1분기 월평균 : 1,2,3월의 총범죄수 평균)

import numpy as np

cond1 = df['범죄분류'].str.contains('1월|2월|3월', na=False)
cond2 = df['범죄분류'].str.contains('4월|5월|6월', na=False)
cond3 = df['범죄분류'].str.contains('7월|8월|9월', na=False)
cond4 = df['범죄분류'].str.contains('10월|11월|12월', na=False)

choices = [
    df['범죄분류'].str[:4] + '1분기',
    df['범죄분류'].str[:4] + '2분기',
    df['범죄분류'].str[:4] + '3분기',
    df['범죄분류'].str[:4] + '4분기'
]

df['분기'] = np.select([cond1, cond2, cond3, cond4], choices, default=np.nan)

df['총범죄수'] = df['절도']  +df['사기']  +df['배임']  +df['방화']  +df['폭행']
an3 = df[df['분기'] == df.groupby('분기')['총범죄수'].mean().sort_values(ascending=False).reset_index().iloc[0]['분기']]['사기'].max()

print(an3)
```
